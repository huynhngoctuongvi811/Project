3. Thống kê mô tả: các đại lượng thống kê cơ bản như trung bình, trung vị, độ lệch chuẩn, v.v. về chỉ số giá
3.1 Các đại lượng thống kê
- Giá trị trung bình (mean) là một trong những chỉ số thống kê cơ bản, dùng để đo lường trung tâm của một tập hợp dữ liệu. Giá trị trung bình trong khoảng thời gian tính trung bình giá đóng cửa mở cửa cao nhất thấp nhất giá đóng cửa điều chỉnh trong vòng 6 tháng
- Trung vị (median) là một chỉ số thống kê quan trọng, dùng để xác định giá trị giữa của một tập hợp dữ liệu khi đã được sắp xếp theo thứ tự. Trung vị rất hữu ích trong việc đo lường trung tâm của dữ liệu, đặc biệt khi dữ liệu có giá trị ngoại lai (outliers), vì nó không bị ảnh hưởng bởi các giá trị quá cao hoặc quá thấp.
=> Cho thấy sự tăng giảm của cổ phiếu
- Độ lệch chuẩn là một chỉ số thống kê thể hiện mức độ phân tán của một tập hợp dữ liệu so với giá trị trung bình của nó. Đối với chỉ số giá, độ lệch chuẩn giúp nhà đầu tư hiểu rõ hơn về biến động của giá trong một khoảng thời gian nhất định.
=> Độ lệch chuẩn càng cao thì sự rủi ro càng lớn và ngược lại
-Biểu đồ Histogram : có thể dễ dàng nhận diện được tần suất xuất hiện của các giá trị trong một tập hợp dữ liệu, thể hiện rõ sự biến động giá cổ phiếu để đánh giá được rủi ro và khả năng sinh lợi nhuận.
- Phân Phối Dữ Liệu: Xem cách dữ liệu phân bố: có đồng đều, tập trung vào một giá trị, hay có nhiều giá trị ngoại lai?
- Đỉnh (Peaks): Các đỉnh trong histogram cho thấy các giá trị thường gặp nhất.
- Tính Đối Xứng: Nếu histogram có dạng đối xứng, dữ liệu có thể phân phối chuẩn. Nếu lệch về bên trái hoặc bên phải, có thể có sự thiên lệch trong dữ liệu.
=> Histogram là một công cụ mạnh mẽ để trực quan hóa phân phối của dữ liệu, giúp hiểu rõ hơn về các đặc điểm của tập dữ liệu
-Biểu đồ Boxplot: là một công cụ trực quan hóa dữ liệu phổ biến trong thống kê, dùng để mô tả phân phối của một biến định lượng và giúp phát hiện các giá trị ngoại lai (outliers). Việc sử dụng boxplot trong phân tích dữ liệu sẽ cung cấp cái nhìn tổng quát và sâu sắc hơn về đặc điểm của dữ liệu.
=> Boxplot rất hữu ích để so sánh phân phối của một biến giữa nhiều nhóm.
3.2. Phân tích phân phối dữ liệu
3.2.1. Hình Dạng Phân Phối
- Đối Xứng (Symmetric): Dữ liệu có thể phân phối đều xung quanh giá trị trung bình, như phân phối chuẩn.
- Lệch (Skewed): Dữ liệu có thể lệch về bên trái (negative skew) hoặc bên phải (positive skew). Điều này có thể cho thấy sự hiện diện của các giá trị ngoại lai.
- Đỉnh (Peakedness): Hình dạng của phân phối có thể cho thấy có nhiều đỉnh (multimodal) hoặc chỉ một đỉnh (unimodal).
3.2.2. Tính Tập Trung (Central Tendency)
- Giá Trị Trung Bình (Mean): Trung bình của tất cả các giá trị trong tập dữ liệu.
- Trung Vị (Median): Giá trị ở giữa khi dữ liệu được sắp xếp. Trung vị không bị ảnh hưởng bởi các giá trị ngoại lai.
- Thống Kê Mode: Giá trị xuất hiện nhiều nhất trong tập dữ liệu.
3.2.3. Phân Tán (Dispersion)
- Độ Lệch Chuẩn (Standard Deviation): Đo lường mức độ phân tán của dữ liệu xung quanh giá trị trung bình.
- Phạm Vi (Range): Khoảng cách giữa giá trị lớn nhất và nhỏ nhất.
- Tứ Phân (Quartiles): Chia dữ liệu thành bốn phần bằng nhau, giúp bạn hiểu hơn về sự phân bố.
3.2.4. Phân Tích Ngoại Lai (Outliers)
- Xác Định Ngoại Lai: Sử dụng các phương pháp như box plot để phát hiện các giá trị ngoại lai, có thể gây ảnh hưởng đến các chỉ số thống kê.
- Phân Tích Tác Động: Đánh giá xem các giá trị ngoại lai có nên được loại bỏ hay giữ lại.
3.2.5. Trực Quan Hóa Dữ Liệu
- Biểu Đồ Histogram: Hiển thị tần suất của các giá trị trong các khoảng.
- Box Plot: Cho phép bạn xem các phần trăm của dữ liệu, giá trị trung vị và các giá trị ngoại lai.
- Biểu Đồ Phân Phối: Hiển thị hình dạng phân phối của dữ liệu.
3.3.Ước lượng và kiểm định là hai kỹ thuật quan trọng trong thống kê
3.3.1. Ước Lượng (Estimation)
Ước lượng là quá trình sử dụng thông tin từ mẫu dữ liệu để ước tính các tham số của tổng thể.
• Ước Lượng Điểm (Point Estimation)
- Sử dụng một giá trị duy nhất từ mẫu để ước lượng tham số của tổng thể
- Ví Dụ: Tính trung bình của một mẫu giá để ước lượng trung bình giá của toàn bộ sản phẩm. • Ước Lượng Khoảng (Interval Estimation)
- Cung cấp một khoảng giá trị (confidence interval) mà trong đó tham số tổng thể có thể nằm - Cách Tính: Tính khoảng tin cậy với độ tin cậy nhất định (thường là 95% hoặc 99%).
3.3.2. Kiểm Định (Hypothesis Testing)
Kiểm định là quá trình đưa ra quyết định về một giả thuyết nào đó dựa trên dữ liệu mẫu.
Các Bước Kiểm Định
- Xác định giả thuyết:
o Giả thuyết không (H0): Giả thuyết muốn kiểm tra
o Giả thuyết thay thế (H1): Giả thuyết nghi ngờ là đúng.
- Mức ý nghĩa (α):
o Thường là 0.05 hoặc 0.01.
- Tính toán thống kê kiểm định:
o Sử dụng các phương pháp như t-test, chi-squared test, ANOVA tùy thuộc vào
loại dữ liệu và giả thuyết. -Ra quyết định:
o So sánh giá trị p với α:
• Nếu p < α: bác bỏ H0, chấp nhận H1
• Nếu p ≥ α: không có đủ bằng chứng bác bỏ H0, giữ nguyên H0
=> Ước lượng và kiểm định là các kỹ thuật quan trọng giúp chúng ta đưa ra kết luận dựa trên dữ liệu. Việc hiểu rõ hai quá trình này sẽ thực hiện phân tích thống kê một cách hiệu quả hơn.

o Ma trận tương quan :
Mục đích:
Ma trận tương quan cho phép ta đánh giá mức độ tương quan tuyến tính giữa các cặp cổ phiếu (giá hoặc lợi nhuận). Ma trận tương quan cung cấp hệ số tương quan giữa tất cả các cặp biến trong tập dữ liệu. Trong phân tích dữ liệu tài chính, ma trận tương quan (correlation matrix) đóng vai trò quan trọng trong việc hiểu mối quan hệ giữa các biến số. Đặc biệt, khi phân tích giá cổ phiếu, việc nắm bắt tương quan giữa các giá trị như giá mở cửa (open), giá cao nhất (high), giá thấp nhất (low), giá đóng cửa (close), và khối lượng giao dịch (volume) có thể cung cấp thông tin quan trọng về xu hướng thị trường và sự biến động của giá cổ phiếu. Mức tương quan được đo bằng hệ số tương quan Pearson, giá trị này nằm trong khoảng [-1, 1].
=> Kết quả đạt được:

- Tương quan giữa giá cao nhất và giá thấp nhất: Cũng thường có hệ số dương, cho thấy

- Tương quan giữa khối lượng giao dịch và giá: Có thể có hệ số thấp hoặc vừa, tùy

Ma trận tương quan giúp xác định và đánh giá mối quan hệ giữa các biến trong tập dữ liệu. Nếu có sự tương quan mạnh giữa các biến, có thể hỗ trợ nhà đầu tư trong việc đánh giá, xem xét và quyết định đầu tư.
4.2.Hệ số tương quan (Correlation) và độ đồng biến (Covariance)
- Hệ số tương quan: Hệ số Pearson là một chỉ số cho biết mức độ liên quan giữa hai biến, thường được dùng để đánh giá mối quan hệ tuyến tính.
- Độ đồng biến (Covariance): Là một chỉ số đo lường mức độ thay đổi đồng thời giữa hai biến số. Nếu giá trị dương, hai biến có xu hướng cùng chiều; nếu âm, chúng có xu hướng ngược chiều.
4.3 Kết quả của hệ số tương quan và độ đồng biến
Correlation:
• Hệ số tương quan gần 1 cho thấy mối liên hệ thuận hoàn hảo giữa hai biến. Khi một biến tăng thì biến kia cũng tăng theo.
• Hệ số tương quan gần -1 cho thấy mối liên hệ đối nghịch hoàn hảo giữa hai biến. Khi một biến tăng thì biến còn lại sẽ giảm.
• Hệ số tương quan gần 0 cho thấy mối liên hệ yếu hoặc không có liên hệ giữa hai biến. Covariance:
• Đồng biến dương: khi một biến tăng, biến kia cũng tăng.
• Đồng biến âm: khi một biến tăng, biến kia giảm.
• Đồng biến bằng 0: không có mối liên hệ rõ ràng

####2.3.3 Xử lý giá trị ngoại lai


```r

detect_outliers <- function(x) {
  q1 <- quantile(x, 0.25)
  q3 <- quantile(x, 0.75)
  iqr <- q3 - q1
  lower_bound <- q1 - 1.5 * iqr
  upper_bound <- q3 + 1.5 * iqr
  x[x < lower_bound | x > upper_bound]
}

# Thay thế giá trị ngoại lai bằng NA
handle_outliers <- function(df, col_names) {
  for (col_name in col_names) {
    outliers <- detect_outliers(df[[col_name]])
    if (length(outliers) > 0) {
      df[[col_name]][df[[col_name]] %in% outliers] <- NA
    }
  }
  return(df)
}

# Xử lý ngoại lai cho df_HPG
# Ví dụ: Xử lý ngoại lai cho các cột 'open', 'high', 'low', 'close', 'volume'
numeric_cols_hpg <- c("open", "high", "low", "close", "volume")
df_HPG <- handle_outliers(df_HPG, numeric_cols_hpg)


# Xử lý ngoại lai cho df_VIC
numeric_cols_vic <- c("open", "high", "low", "close", "volume")
df_VIC <- handle_outliers(df_VIC, numeric_cols_vic)

# Xử lý ngoại lai cho df_VCG
numeric_cols_vcg <- c("open", "high", "low", "close", "volume")
df_VCG <- handle_outliers(df_VCG, numeric_cols_vcg)

sum(is.na(df_HPG))
```

```r

# Impute missing values with the mean for numeric columns
impute_missing_values <- function(df) {
  for (col in names(df)) {
    if (is.numeric(df[[col]])) {
      df[[col]][is.na(df[[col]])] <- mean(df[[col]], na.rm = TRUE)
    }
  }
  return(df)
}

df_HPG <- impute_missing_values(df_HPG)
df_VCG <- impute_missing_values(df_VCG)
df_VIC <- impute_missing_values(df_VIC)
```

Biểu đồ cung cấp thông tin rất hữu ích để phân tích và đưa ra quyết định về sự ổn định và tiềm năng của các công ty trong tương lai.

# 4.Ước lượng và kiểm định


```r
# Dữ liệu giả định cho Hoà Phát (High và Low)
data <- data.frame(
  price_type = factor(rep(c("high", "low"), each = 6)),  # Tạo biến nhóm
  price = c(13.97, 14.12, 13.87, 13.16, 13.21, 12.91,  # Giá cao nhất
            13.52, 13.62, 13.01, 12.86, 12.76, 12.05)    # Giá thấp nhất
)

# Tính toán thống kê cơ bản
statistics <- data.frame(
  mean = sapply(split(data$price, data$price_type), mean), # Tính mean theo nhóm
  median = sapply(split(data$price, data$price_type), median), # Tính median theo nhóm
  std_dev = sapply(split(data$price, data$price_type), sd), # Tính sd theo nhóm
  min = sapply(split(data$price, data$price_type), min), # Tính min theo nhóm
  max = sapply(split(data$price, data$price_type), max) # Tính max theo nhóm
)
rownames(statistics) <- levels(data$price_type) # Đặt tên hàng cho statistics

# Kiểm định t hai mẫu Welch (so sánh High và Low)
t_test_result <- t.test(price ~ price_type, data = data) # Sử dụng formula interface

# Kiểm định Bartlett (đồng nhất phương sai)
bartlett_test_result <- bartlett.test(price ~ price_type, data = data) # Sử dụng formula interface

# In kết quả
print(statistics)
print(t_test_result)
print(bartlett_test_result)
```

## Biểu đồ Histogram

**Biểu đồ histogram**: là một công cụ trực quan hóa dữ liệu dưới dạng các cột biểu diễn tần suất xuất hiện của một biến số trong các khoảng giá trị (bins) nhất định.

```r
# Tạo một cột để nhóm 5 năm
df_HPG <- df_HPG %>%
  mutate(year = as.numeric(format(time, "%Y")),
         year_group = cut(year, breaks = seq(2000, 2025, by = 5), include.lowest = TRUE, right = FALSE))

df_VCG <- df_VCG %>%
  mutate(year = as.numeric(format(time, "%Y")),
         year_group = cut(year, breaks = seq(2000, 2025, by = 5), include.lowest = TRUE, right = FALSE))

df_VIC <- df_VIC %>%
  mutate(year = as.numeric(format(time, "%Y")),
         year_group = cut(year, breaks = seq(2000, 2025, by = 5), include.lowest = TRUE, right = FALSE))

# Hàm vẽ histogram theo nhóm 5 năm
plot_histogram_by_group <- function(data, variable, title, binwidth = 10) {
  ggplot(data, aes_string(x = variable, fill = "year_group")) +
    geom_histogram(binwidth = binwidth, color = "black", alpha = 0.7, position = "identity") +
    scale_fill_brewer(palette = "Set2", name = "Khoảng năm")+
    labs(title = title, x = variable, y = "Tần suất") +
    theme_minimal() +
    theme(axis.text.x = element_text(angle = 45, hjust = 1))
}

# Vẽ histogram cho các biến
hist_HPG_close_by_group <- plot_histogram_by_group(df_HPG, "close", "HPG: Phân phối giá đóng cửa theo nhóm 5 năm", binwidth = 5)
hist_VCG_close_by_group <- plot_histogram_by_group(df_VCG, "close", "VCG: Phân phối giá đóng cửa theo nhóm 5 năm", binwidth = 5)
hist_VIC_close_by_group <- plot_histogram_by_group(df_VIC, "close", "VIC: Phân phối giá đóng cửa theo nhóm 5 năm", binwidth = 5)

# Hiển thị biểu đồ
hist_HPG_close_by_group
hist_VCG_close_by_group
hist_VIC_close_by_group
```

```r
# Tính ma trận tương quan cho từng mã cổ phiếu
corr_HPG <- cor(df_HPG[, c("open", "high", "low", "close", "volume")])
corr_VCG <- cor(df_VCG[, c("open", "high", "low", "close", "volume")])
corr_VIC <- cor(df_VIC[, c("open", "high", "low", "close", "volume")])

# Vẽ heatmap cho từng mã cổ phiếu
plot_heatmap <- function(corr_matrix, title) {
  corr_melt <- melt(corr_matrix)
  ggplot(data = corr_melt, aes(Var1, Var2, fill = value)) +
    geom_tile(color = "white") +
    scale_fill_gradient2(low = "blue", mid = "white", high = "red", midpoint = 0,
                         limit = c(-1, 1), space = "Lab", name = "Tương quan") +
    geom_text(aes(label = round(value, 2)), color = "black", size = 4) +
    labs(title = title, x = "", y = "") +
    theme_minimal() +
    theme(axis.text.x = element_text(angle = 45, hjust = 1))
}

# Heatmap cho từng mã
heatmap_HPG <- plot_heatmap(corr_HPG, "Tương quan giữa các biến - HPG")
heatmap_VCG <- plot_heatmap(corr_VCG, "Tương quan giữa các biến - VCG")
heatmap_VIC <- plot_heatmap(corr_VIC, "Tương quan giữa các biến - VIC")

# Hiển thị các biểu đồ
heatmap_HPG
heatmap_VCG
heatmap_VIC
```
